### 텐서에 수학 연산 적용하기

```python
tf.random.set_seed(1)

t1 = tf.random.uniform(shape=(5, 2), #[-1,1) 사이의 균등 분포
                       minval=-1.0,
                       maxval=1.0)

t2 = tf.random.normal(shape=(5, 2), #표준 정규 분포
                      mean=0.0,
                      stddev=1.0)
```

```python
t3 = tf.multiply(t1, t2).numpy() #원소별 곱셈
print(t3)
```

```python
t4 = tf.math.reduce_mean(t1, axis=0) #평균

print(t4)
```

```python
t5 = tf.linalg.matmul(t1, t2, transpose_b=True) #행렬곱셈(t2 전치)

print(t5.numpy())
```

```python
t6 = tf.linalg.matmul(t1, t2, transpose_a=True) #행렬곱셈(t1 전치)

print(t6.numpy())
```

```python
norm_t1 = tf.norm(t1, ord=2, axis=1).numpy() #노름(=벡터의 크기) 계산

print(norm_t1)
```

### split(), stack(), concatenate() 함수

```python
#분할 개수 지정
tf.random.set_seed(1)

t = tf.random.uniform((6,)) #0과 1 사이 중 난수 6개 생성

print(t.numpy())

t_splits = tf.split(t, 3) #크기 6인 텐서를 크기 2인 텐서 세개로 나눔

[item.numpy() for item in t_splits] #각 부분을 numpy 배열로 변환
```

```python
#다른 분할 크기 전달
tf.random.set_seed(1)
t = tf.random.uniform((5,))

print(t.numpy())

t_splits = tf.split(t, num_or_size_splits=[3, 2]) #크기 5인 텐서를 크기가 3하고 2인 텐서로 나누기

[item.numpy() for item in t_splits]
```

```python
A = tf.ones((3,))
B = tf.zeros((2,))

C = tf.concat([A, B], axis=0) # 1D로 합치기
print(C.numpy())
```

```python
A = tf.ones((3,))
B = tf.zeros((3,))

S = tf.stack([A, B], axis=1) # 2D로 합치기
print(S.numpy())
```

## 텐서플로 데이터셋 API(tf.data)를 사용하여 입력 파이프라인 구축하기

### 텐서에서 텐서플로 데이터셋 만들기

```python
a = [1.2, 3.4, 7.5, 4.1, 5.0, 1.0]

ds = tf.data.Dataset.from_tensor_slices(a)

print(ds)
```

```python
#개별 원소를 순환
for item in ds:
    print(item)
```

```python
ds_batch = ds.batch(3) #크기로 나누기

for i, elem in enumerate(ds_batch, 100):
    print('batch {}:'.format(i), elem.numpy())
```

### 두 개의 텐서를 하나의 데이터셋으로 연결하기

```python
tf.random.set_seed(1)

t_x = tf.random.uniform([4, 3], dtype=tf.float32) #4x3 혇태, 0~1 난수 값
t_y = tf.range(4) #0~3까지의 정수 값 4개
```

```python
#1:1 대응, 방법1
ds_x = tf.data.Dataset.from_tensor_slices(t_x) 
ds_y = tf.data.Dataset.from_tensor_slices(t_y)

ds_joint = tf.data.Dataset.zip((ds_x, ds_y)) #합치기

for example in ds_joint:
    print('  x: ', example[0].numpy(),
          '  y: ', example[1].numpy())
```

```python
## 방법 2:
ds_joint = tf.data.Dataset.from_tensor_slices((t_x, t_y))

for example in ds_joint:
    print('  x: ', example[0].numpy(),
          '  y: ', example[1].numpy())
```

```python

ds_trans = ds_joint.map(lambda x, y: (x*2-1.0, y))

for example in ds_trans:
    print('  x: ', example[0].numpy(),
          '  y: ', example[1].numpy())
```

### shuffle(), batch(), repeat() 메서드

```python
tf.random.set_seed(1)
ds = ds_joint.shuffle(buffer_size=len(ds_x))#행 섞기

for example in ds:
    print('  x: ', example[0].numpy(),
          '  y: ', example[1].numpy())
```

```python
ds = ds_joint.batch(batch_size=3, #배치로 나누기
                    drop_remainder=False)

batch_x, batch_y = next(iter(ds))

print('배치 x: \n', batch_x.numpy())

print('배치 y:   ', batch_y.numpy())
```

```python
ds = ds_joint.batch(3).repeat(count=2) #3개로 나뉘고 2번 반복

for i,(batch_x, batch_y) in enumerate(ds):
    print(i, batch_x.shape, batch_y.numpy()) 
```

```python
ds = ds_joint.repeat(count=2).batch(3) #2번 반복하는데 3개로 나누기

for i,(batch_x, batch_y) in enumerate(ds):
    print(i, batch_x.shape, batch_y.numpy())
```

```python
tf.random.set_seed(1)

## 순서 1: shuffle -> batch -> repeat
ds = ds_joint.shuffle(4).batch(2).repeat(3)

for i,(batch_x, batch_y) in enumerate(ds):
    print(i, batch_x.shape, batch_y.numpy())
```

```python
tf.random.set_seed(1)

## 순서 1: shuffle -> batch -> repeat
ds = ds_joint.shuffle(4).batch(2).repeat(20)

for i,(batch_x, batch_y) in enumerate(ds):
    print(i, batch_x.shape, batch_y.numpy())
```

```python
tf.random.set_seed(1)

## 순서 2: batch -> shuffle -> repeat
ds = ds_joint.batch(2).shuffle(4).repeat(3)

for i,(batch_x, batch_y) in enumerate(ds):
    print(i, batch_x.shape, batch_y.numpy())
```

```python
tf.random.set_seed(1)

## 순서 2: batch -> shuffle -> repeat
ds = ds_joint.batch(2).shuffle(4).repeat(20)

for i,(batch_x, batch_y) in enumerate(ds):
    print(i, batch_x.shape, batch_y.numpy())
```

```python
tf.random.set_seed(1)

## 순서 3: batch -> repeat -> shuffle
ds = ds_joint.batch(2).repeat(3).shuffle(4)

for i,(batch_x, batch_y) in enumerate(ds):
    print(i, batch_x.shape, batch_y.numpy())
```

### 로컬 디스크에 있는 파일에서 데이터셋 만들기

```python
import pathlib

imgdir_path = pathlib.Path('cat_dog_images')

file_list = sorted([str(path) for path in imgdir_path.glob('*.jpg')])

print(file_list)
```

```python
import matplotlib.pyplot as plt
import os


fig = plt.figure(figsize=(10, 5))
for i,file in enumerate(file_list):
    img_raw = tf.io.read_file(file)
    img = tf.image.decode_image(img_raw)
    print('이미지 크기: ', img.shape)
    ax = fig.add_subplot(2, 3, i+1)
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(img)
    ax.set_title(os.path.basename(file), size=15)

plt.tight_layout()
plt.show()
```

```python
labels = [1 if 'dog' in os.path.basename(file) else 0 #강아지 1, 고양이 0
          for file in file_list]
print(labels)
```

```python
ds_files_labels = tf.data.Dataset.from_tensor_slices(
    (file_list, labels)) #2개 합치기 위해 실행함 (파일 경로, 라벨)

for item in ds_files_labels:
    print(item[0].numpy(), item[1].numpy())
```

```python
#원본 이미지를 로드하고 디코딩한 후 이미지 크기 변경 -> 데이터셋 반환
def load_and_preprocess(path, label): 
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [img_height, img_width])
    image /= 255.0

    return image, label

img_width, img_height = 120, 80

ds_images_labels = ds_files_labels.map(load_and_preprocess) #map : 람다 함수 적용

fig = plt.figure(figsize=(10, 5))
for i,example in enumerate(ds_images_labels):
    print(example[0].shape, example[1].numpy())
    ax = fig.add_subplot(2, 3, i+1)
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(example[0])
    ax.set_title('{}'.format(example[1].numpy()),
                 size=15)

plt.tight_layout()
plt.show()
```

데이터셋을 로드하는 또 다른 방법입니다.

```python
mnist, mnist_info = tfds.load('mnist', with_info=True,
                              shuffle_files=False)

print(mnist_info)

print(mnist.keys())
```

```python
ds_train = mnist['train']

assert isinstance(ds_train, tf.data.Dataset)

ds_train = ds_train.map(lambda item:
     (item['image'], item['label']))

ds_train = ds_train.batch(10)
batch = next(iter(ds_train))
print(batch[0].shape, batch[1])

fig = plt.figure(figsize=(15, 6))
for i,(image,label) in enumerate(zip(batch[0], batch[1])):
    ax = fig.add_subplot(2, 5, i+1)
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(image[:, :, 0], cmap='gray_r')
    ax.set_title('{}'.format(label), size=15)

# plt.savefig('images/13_4.png', dpi=300)
plt.show()
```
